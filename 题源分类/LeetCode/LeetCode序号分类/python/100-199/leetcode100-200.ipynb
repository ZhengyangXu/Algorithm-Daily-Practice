```python
#
# @lc app=leetcode.cn id=108 lang=python3
#
# [108] 将有序数组转换为二叉搜索树
#
# https://leetcode-cn.com/problems/convert-sorted-array-to-binary-search-tree/description/
#
# algorithms
# Easy (74.50%)
# Likes:    655
# Dislikes: 0
# Total Accepted:    128.6K
# Total Submissions: 172.5K
# Testcase Example:  '[-10,-3,0,5,9]'
#
# 将一个按照升序排列的有序数组，转换为一棵高度平衡二叉搜索树。
# 
# 本题中，一个高度平衡二叉树是指一个二叉树每个节点 的左右两个子树的高度差的绝对值不超过 1。
# 
# 示例:
# 
# 给定有序数组: [-10,-3,0,5,9],
# 
# 一个可能的答案是：[0,-3,9,-10,null,5]，它可以表示下面这个高度平衡二叉搜索树：
# 
# ⁠     0
# ⁠    / \
# ⁠  -3   9
# ⁠  /   /
# ⁠-10  5
# 
# 
#

# @lc code=start
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

def buildTree(nums):
    if not nums:
        return  

    size = len(nums) - 1
    mid = size // 2 
    root = TreeNode(nums[mid])
    print("root:%s"%root.val)
    root.left = buildTree(nums[:mid])
    print("root:%s,root.left:%s"%(root.val,root.left.val) if root.left and root else "root:%s,root.left:%s"%(root.val,None))
    root.right = buildTree(nums[mid+1:])
    print("root:%s,root.right:%s"%(root.val,root.right.val) if root.right and root else "root:%s,root.right:%s"%(root.val,None))

    return root
```

```python
nums = [-10,-3,0,5,9]
```

```python
x = buildTree(nums)
```
